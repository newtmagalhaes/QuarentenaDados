## Desafio 2

No gráfico de budget por profit temos um ponto com muito custo e prejuizo, descubra com é esse filme (budget próximo de 2.5).

```python
# Adicionando coluna de lucro (profit)
imdbUSA['profit'] = imdbUSA['gross'] - imdbUSA['budget']

# Gráfico Custo/Lucro (budget/profit)
sns.scatterplot(
    data=imdbUSA,
    x='budget',
    y='profit'
)
```

```python
imdbUSA.sort_values(by='profit')[['movie_title', 'title_year', 'profit']].head()
```
